```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
ncbi_pyxi_length_txt = pd.read_csv("/Users/guangyichen/Desktop/WIBI/Master-Thesis/Data/Final-Data/all_length.txt", header = None, delim_whitespace=True)
ncbi_pyxi_length = ncbi_pyxi_length_txt[0]

# set figure size
from matplotlib.pyplot import figure
figure(figsize=(10, 8), dpi=300)

plt.hist(ncbi_pyxi_length, bins='auto', color ="#69b3a2")
plt.title("All length distribution (71 clades)", fontweight ='bold', fontsize = 13)
plt.xlabel("bp", fontweight ='bold', fontsize = 13)
plt.ylabel("counts", fontweight ='bold', fontsize = 13)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.xlim([8000000,15000000])
plt.show()
```
